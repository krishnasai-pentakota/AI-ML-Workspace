##Conditional Statement
##Basically it is a set of statements which will work based on conditions like whether it is true or not.

```python
##If statement
a = 10
if a > 5:
    print("a is greater than 5") ##if statement is used to check a condition and execute a block of code if the condition is true. In this case, since a is 10, which is greater than 5, the message "a is greater than 5" will be printed.
```

```python
##If-Else statement
a = 3
if a > 5:    ##one note is that the statements actually automatically detects the indentation.
    print("a is greater than 5")
else:
    print("a is not greater than 5") ##else statement is used to execute a block of code if the condition in the if statement is false. In this case, since a is 3, which is not greater than 5, the message "a is not greater than 5" will be printed.
```

##elif statements
##elif statement is used to check for multiple conditions at a single time,

```python
age = int(input("Enter your age: "))
if age <14:
    print("You are a child.")
elif age>14 and age<=18:           ##here, first if block fails breaks or false, it will check the elif block, ## if it is true then it will execute the block of code inside it. If it is false then it will check the else block.
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
num_1 = int(input("Enter first number: "))
num_2 = int(input("Enter second number: "))
operator = input("Enter an operator (+, -, *, /, //, %, **): ")
if operator == "+":
    result = num_1 + num_2
    print(f"The result of {num_1} + {num_2} is: {result}")
elif operator == "-":
    result = num_1 - num_2
    print(f"The result of {num_1} - {num_2} is: {result}")
elif operator == "*":
    result = num_1 * num_2
    print(f"The result of {num_1} * {num_2} is: {result}")                              

elif operator == "/":
    result = num_1 / num_2
    print(f"The result of {num_1} / {num_2} is: {result}")
elif operator == "//":
    result = num_1 // num_2
    print(f"The result of {num_1} // {num_2} is: {result}")
elif operator == "%":
    result = num_1 % num_2
    print(f"The result of {num_1} % {num_2} is: {result}")
elif operator == "**":
    result = num_1 ** num_2
    print(f"The result of {num_1} ** {num_2} is: {result}")
else:
    print("Invalid operator!")
```
