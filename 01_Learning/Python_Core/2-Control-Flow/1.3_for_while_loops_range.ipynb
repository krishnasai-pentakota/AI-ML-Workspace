1. for loops is basically used to iterate over  any range like we can use range functions, strings(it can iterate through every character.), any range accepted data structures like  tuples, lists.
Syntax: for i(#called as iterator, this is the variable which iterates through each element) in range:

```python
##we can use simple for loop to print top 10 numbers
for i in range(1, 11):
    print(i)
```

while loop runs on when the condition is true and it can cause infinite loop, if no condition specified to terminate the loop or to make the condition fail after certain iterations.

```python
n=1
while n<10:
    print(n * n)
    n +=1
```

##Loop Control Statements.
Break: it will used to terminate from the loop immediately and stops the iteration on next iterations

```python
num = 2
while True:
    if num % 2 == 0:
        print(num)
    num += 1
    if num > 10:
        break
```

##Continue Statement.
It is used to skip the current iteration and iterate from next elements.

```python
for i in range(1, 11):
    if i== 5:
        continue
    print(i)
```

##Range function
Range Syntax: ( start, stop, steps.), stop value is exclusive and start is inclusive. steps is used to skip based on the step number.

```python
##For else loop is working in python, it is used to execute a block of code when the loop is completed normally without any break statement.
##The else block will be executed after the for loop finishes iterating over all the items in the sequence.
num = int(input("Enter a number: "))
if num >1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```
